```python
import sys
import pandas as pd
import matplotlib.pyplot as plt

MODEL_PATH = "/content/drive/MyDrive/bert-imdb"
PROJECT_ROOT = "/content/nlp-finetuning-bert"
sys.path.append(PROJECT_ROOT)
```

```python
preds_path = f"{PROJECT_ROOT}/data/predictions.csv"
df = pd.read_csv(preds_path)

df.head()
```

```python
plt.figure()
plt.hist(df["confidence"], bins=20)
plt.xlabel("Confidence")
plt.ylabel("Count")
plt.title("Distribution of model confidence")
plt.show()
```

```python
borderline = df[(df["confidence"] > 0.45) & (df["confidence"] < 0.6)]
borderline
```

```python
df.sort_values("confidence", ascending=False).head(10)
```

```python
df.sort_values("confidence", ascending=True).head(10)
```

```python
threshold = 0.7

df["label_thresholded"] = df.apply(
    lambda r: r["label"] if r["confidence"] >= threshold else "UNCERTAIN",
    axis=1
)

df["label_thresholded"].value_counts()
```

```python
df.to_csv(f"{PROJECT_ROOT}/data/predictions_with_analysis.csv", index=False)
```
